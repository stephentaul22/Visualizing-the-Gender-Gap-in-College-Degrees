# Visualizing the Gender Gap in College Degrees
A gender gap often exists in many college degrees. Today I will use data visualization techniques to visualize this gap in various college majors. I will also save the data as a png file for quick future reference.

I will be using data generated by <a link="https://nces.ed.gov/programs/digest/2013menu_tables.asp">The Department of Education Statistics</a>. I will use a cleaned and compiled version of this data found <a link="http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/">here</a>.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

women_degrees = pd.read_csv('/Users/admin/Downloads/DATASCIENCE PROJECTS - GUIDED/Project 5 | Visualizing the Gender Gap in College Degrees/percent-bachelors-degrees-women-usa.csv')

women_degrees.head()
```

```python
fig = plt.figure(figsize=(18, 20))

stem = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
liberal_arts = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

dark_blue = (0/255,107/255,164/255)
orange = (255/255, 128/255, 14/255)
light_gray = (171/255, 171/255, 171/255)

for value in range(0,18,3):
    index = int(value/3)
    ax = fig.add_subplot(6,3,value+1)
    ax.plot(women_degrees['Year'], women_degrees[stem[index]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem[index]], c=orange, label='Men', linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=light_gray, alpha=0.3)    

    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    if index == 0:
        ax.text(2005, 82, 'Women')
        ax.text(2007, 14, 'Men')
    elif index == 5:
        ax.text(2007, 87, 'Men')
        ax.text(2005, 7, 'Women')

    if index == 5:
        ax.tick_params(labelbottom=True)

for value in range(0,15,3):
    index = int((value)/3)
    ax = fig.add_subplot(6,3,value+2)
    ax.plot(women_degrees['Year'], women_degrees[liberal_arts[index]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[liberal_arts[index]], c=orange, label='Men', linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(liberal_arts[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=light_gray, alpha=0.3)

    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    if index == 0:
        ax.text(2005, 76, 'Women')
        ax.text(2007, 20, 'Men')

    if index == 4:
        ax.tick_params(labelbottom=True)

for value in range(0,18,3):
    index = int((value)/3)
    ax = fig.add_subplot(6,3,value+3)
    ax.plot(women_degrees['Year'], women_degrees[other[index]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other[index]], c=orange, label='Men', linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=light_gray, alpha=0.3)

    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    if index == 0:
        ax.text(2005, 90, 'Women')
        ax.text(2007, 5, 'Men')
    elif index == 5:
        ax.text(2007, 62, 'Men')
        ax.text(2005, 32, 'Women')

    if index == 5:
        ax.tick_params(labelbottom=True)
        
fig.savefig("visualizing_the_gender_gap_in_college_degrees.png")
plt.show()
```

## Conclusion
Some college majors in each category (STEM, Liberal Arts, and Other) have closed the gender gap, while others continue to have a vast gap. Still, others have reversed the gap from male-favored to female-favored. No major has reversed a gap from being female-favored to male-favored.

1. Majors that have closed the gap after being heavily male-favored are: (STEM) Biology, Math & Statistics, Physical Sciences; (Liberal Arts) Communications & Journalism, Social Sciences & History; and (Other) Agriculture, Business, and Architecture.


2. Majors that continue to have a wide male-favored gender gap are: (STEM) Computer Science and Engineering.


3. The major that most significantly reversed the gap from male-favored to female-favored is (STEM) Psychology. The gap also reversed like this in (STEM) Biology and (Liberal Arts) Communications & Journalism, but the gap for these majors is now very small.


4. Majors that have had and continue to have a vast female-favored gap are: (Other) Health Professions, Public Administration, and Education. The majors (Liberal Arts) Foreign Language, English, and Art & Peformance also favor women, but their gender gaps a very small.


5. Most liberal arts majors have had and continue to have a very small gender gap that usually favors women. The only exception for this rule is the Social Sciences & History major, which has virtually eliminated its gap entirely (50/50 Male/Female).
