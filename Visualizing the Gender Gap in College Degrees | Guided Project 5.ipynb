# Visualizing the Gender Gap in College Degrees
A gender gap often exists in many college degrees. Today I will use data visualization techniques to visualize this gap in various college majors.

I will be using data generated by <a link="https://nces.ed.gov/programs/digest/2013menu_tables.asp">The Department of Education Statistics</a>. A student of the University of Pennsylvania, Randal Olsen, has cleaned and compiled some of this data and, it is available on his website <a link="http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/">here</a>. This is the dataset I will be using.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

women_degrees = pd.read_csv('/Users/admin/Downloads/DATASCIENCE PROJECTS - GUIDED/Project 5 | Visualizing the Gender Gap in College Degrees/percent-bachelors-degrees-women-usa.csv')

women_degrees.head()
```

```python
fig = plt.figure(figsize=(18, 20))

stem = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
liberal_arts = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

dark_blue = (0/255,107/255,164/255)
orange = (255/255, 128/255, 14/255)
light_gray = (171/255, 171/255, 171/255)

for value in range(0,18,3):
    index = int(value/3)
    ax = fig.add_subplot(6,3,value+1)
    ax.plot(women_degrees['Year'], women_degrees[stem[index]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem[index]], c=orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=light_gray, alpha=0.3)
    
    if index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
    if index == 5:
        ax.tick_params(labelbottom=True)

for value in range(0,15,3):
    index = int((value)/3)
    ax = fig.add_subplot(6,3,value+2)
    ax.plot(women_degrees['Year'], women_degrees[liberal_arts[index]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[liberal_arts[index]], c=orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(liberal_arts[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=light_gray, alpha=0.3)  
    
    if index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    if index == 4:
        ax.tick_params(labelbottom=True)

for value in range(0,18,3):
    index = int((value)/3)
    ax = fig.add_subplot(6,3,value+3)
    ax.plot(women_degrees['Year'], women_degrees[other[index]], c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other[index]], c=orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=light_gray, alpha=0.3)
    
    if index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
    if index == 5:
        ax.tick_params(labelbottom=True)
        
fig.savefig("visualizing_the_gender_gap_in_college_degrees.png")
plt.show()
```
